```python
"""
Simple Audio Violence Detection Training - Guaranteed to Work
Shelid Vision - Sexual Violence Detection for Women's Safety

This version will work in any environment (Kaggle, Colab, local)
"""

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

def create_synthetic_audio_dataset():
    """
    Create synthetic audio features dataset for training
    This simulates real audio features for violence detection
    """
    print("üîß Creating synthetic audio dataset...")
    
    # Dataset parameters
    n_samples = 5000  # Increased for better training
    n_features = 200  # Comprehensive audio features
    
    # Violence categories
    categories = {
        'normal': 0,      # Normal, calm speech
        'distress': 1,    # Fear, sadness, distress
        'anger': 2,       # Anger, aggression
        'violence': 3,    # High-intensity violence
        'panic': 4        # Extreme distress, panic
    }
    
    # Generate realistic audio features
    np.random.seed(42)
    
    # Create different feature patterns for each category
    data = []
    labels = []
    
    for category, label in categories.items():
        if category == 'normal':
            # Normal speech: low energy, stable features
            n_cat = 2000
            features = np.random.normal(0, 0.5, (n_cat, n_features))
            # Add some structure
            features[:, :50] = np.random.normal(0, 0.3, (n_cat, 50))  # MFCC-like
            
        elif category == 'distress':
            # Distress: higher pitch variation, trembling
            n_cat = 1200
            features = np.random.normal(0.3, 0.8, (n_cat, n_features))
            # Higher frequency components
            features[:, 50:100] = np.random.normal(0.5, 1.0, (n_cat, 50))
            
        elif category == 'anger':
            # Anger: high energy, harsh tones
            n_cat = 1000
            features = np.random.normal(0.5, 1.2, (n_cat, n_features))
            # High energy features
            features[:, 100:150] = np.random.normal(1.0, 1.5, (n_cat, 50))
            
        elif category == 'violence':
            # Violence: chaotic, sudden energy spikes
            n_cat = 500
            features = np.random.normal(1.0, 2.0, (n_cat, n_features))
            # Chaotic patterns
            features[:, 150:200] = np.random.normal(1.5, 2.5, (n_cat, 50))
            
        elif category == 'panic':
            # Panic: extreme values, very high variation
            n_cat = 300
            features = np.random.normal(1.5, 3.0, (n_cat, n_features))
            # Extreme variations
            features += np.random.exponential(0.5, (n_cat, n_features))
        
        data.extend(features)
        labels.extend([label] * n_cat)
    
    # Convert to numpy arrays
    X = np.array(data)
    y = np.array(labels)
    
    # Shuffle the dataset
    indices = np.random.permutation(len(X))
    X = X[indices]
    y = y[indices]
    
    print(f"‚úÖ Synthetic dataset created: {X.shape[0]} samples, {X.shape[1]} features")
    print(f"üìä Label distribution:")
    for cat, label in categories.items():
        count = np.sum(y == label)
        print(f"   {cat.capitalize()}: {count} samples ({count/len(y)*100:.1f}%)")
    
    return X, y

def build_audio_violence_model(input_shape):
    """Build neural network for audio violence detection"""
    model = Sequential([
        # Input layer
        Dense(512, input_shape=(input_shape,), activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        
        # Hidden layers
        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.4),
        
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        
        Dense(32, activation='relu'),
        Dropout(0.2),
        
        # Output layer (5 classes)
        Dense(5, activation='softmax')
    ])
    
    # Compile model
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

def train_audio_violence_detector():
    """Main training function"""
    print("üõ° SHELID VISION - Audio Violence Detection Training")
    print("=" * 60)
    print("üéØ Specialized for Sexual Violence Detection Against Women")
    print("=" * 60)
    
    # Create dataset
    X, y = create_synthetic_audio_dataset()
    
    # Normalize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Split dataset
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42, stratify=y
    )
    
    X_train, X_val, y_train, y_val = train_test_split(
        X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
    )
    
    print(f"üìä Training set: {X_train.shape}")
    print(f"üìä Validation set: {X_val.shape}")
    print(f"üìä Test set: {X_test.shape}")
    
    # Build model
    model = build_audio_violence_model(X_train.shape[1])
    print("\nüèó Model Architecture:")
    model.summary()
    
    # Callbacks
    callbacks = [
        EarlyStopping(monitor='val_accuracy', patience=15, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=8, min_lr=1e-7)
    ]
    
    # Train model
    print("\nüöÄ Starting training...")
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=50,  # Reduced for faster training
        batch_size=32,
        callbacks=callbacks,
        verbose=1
    )
    
    # Evaluate model
    print("\nüìà Evaluating model...")
    test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
    
    # Predictions
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    # Results
    print(f"\nüéØ FINAL RESULTS:")
    print(f"Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
    print(f"Test Loss: {test_loss:.4f}")
    
    # Classification report
    class_names = ['Normal', 'Distress', 'Anger', 'Violence', 'Panic']
    print("\nüìä Classification Report:")
    print(classification_report(y_test, y_pred_classes, target_names=class_names))
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred_classes)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
               xticklabels=class_names, yticklabels=class_names)
    plt.title('Shelid Vision Audio Violence Detection - Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
    
    # Plot training history
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Save model
    model.save('shelid_vision_audio_detector.h5')
    print("\nüíæ Model saved as 'shelid_vision_audio_detector.h5'")
    
    print("\n‚úÖ TRAINING COMPLETED SUCCESSFULLY!")
    print("üéØ Model ready for Shelid Vision deployment!")
    print("üõ° Protecting women through AI technology!")
    
    return model, history, scaler

# Run training
if __name__ == "__main__":
    model, history, scaler = train_audio_violence_detector()
```
